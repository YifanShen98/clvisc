# Distribution of sampled hadrons

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def rapidity(p0, p3):
    return 0.5 * np.log((p0 + p3)/(p0 - p3))

def create_image(event_id=0):
    mc_particle_list = np.loadtxt('event_ideal/mc_particle_list%s'%event_id)
    mass = mc_particle_list[:, 4]
    four_momentum = mc_particle_list[:, 5:9]
    pid = mc_particle_list[:, 9]
    pion_4mom = four_momentum[np.absolute(pid) == 211]
    pion_rapidity = rapidity(pion_4mom[:, 0], pion_4mom[:, 3])
    pion_mid_rapidity = pion_4mom[np.absolute(pion_rapidity) < 1.0]
    a = plt.hist2d(pion_mid_rapidity[:, 1], pion_mid_rapidity[:, 2], bins=70)
```
